```python
import numpy as np, scipy as sp
from matplotlib import pyplot as plt
```

```python
Lx=800;
mu=3;
t=-10;
dt=.5;


# no spatial variation in the hopping vector
# t_x    = t*(np.eye(Lx, Lx,1) + 0*np.eye(Lx, Lx, Lx-1))       
# works for EVEN lattice sizes ATM 

mu_eye = np.eye(Lx, Lx,0) # diagonal vector
t_x    = np.diag([t+dt,t-dt]*(Lx//2-1) + [t+dt], 1) + 0*(t-dt)*np.eye(Lx, Lx, Lx-1) 

H_0 = mu * mu_eye + (t_x + t_x.transpose().conjugate()) 
```

```python
vals, vecs = np.linalg.eigh(H_0)
plt.plot(vals,'.')
plt.show()

plt.plot(vecs[:, 0],'1') #lowest eigenvalue
plt.plot(vecs[:, 1],'2') # second
# plt.plot(vecs[:, 2],'3') # ++
# plt.plot(vecs[:, 3],'4') # ++
# plt.plot(vecs[:, 4],'o') # ++++++
plt.show()
```

```python
fig, axs = plt.subplots(5, 1, figsize=(9,16))

axs[0].plot(range(Lx), vals, '.')
axs[0].set_title('Eigenvalues')

axs[1].plot(range(Lx), vals, '.',)
axs[1].set_title('Zero-mode eigenvalues')
axs[1].set_ylim(2.999, 3.001)
axs[1].set_xlim(390, 410)

axs[2].plot(range(Lx), vecs[:, Lx//2 -1], 'x')
axs[2].plot(range(Lx), vecs[:, Lx//2   ], '.')
# axs[2].plot(range(Lx), vecs[:, Lx//2 +1],'og')
axs[3].set_title('Automatic ticks')

axs[3].plot(range(Lx), vecs[:, 0],'.')
axs[3].plot(range(Lx), vecs[:, 1],'o')
axs[3].set_title('Automatic ticks')


axs[4].plot(range(Lx), vecs[:, 200],'-')
axs[4].plot(range(Lx), vecs[:, 799],'--')
axs[4].set_title('Automatic ticks')

plt.plot()



# axs[1].plot(xdata, data1)
# axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
# axs[1].set_yticks([-1.5, 0, 1.5])  # note that we don't need to specify labels
# axs[1].set_title('Manual ticks');
```

Now we add disorder to break into left/right etc wells.

To really highlight this lets just add a barrier which splits the left and right segments explicitly.

```python
mu_eye = np.diag([0]*(3*Lx//8) + [100]*(Lx//4) + [0]*(3*Lx//8)) # diagonal vector

H_0 = mu * mu_eye + (t_x + t_x.transpose().conjugate()) 
```

```python
vals, vecs = np.linalg.eigh(H_0)
plt.plot(vals,'.')
plt.show()

plt.plot(vecs[:, 601],'1') #lowest eigenvalue
plt.plot(vecs[:, 650],'2') # second
plt.plot(vecs[:, 798],'3') # ++
plt.plot(vecs[:, 799],'4') # ++
# plt.plot(vecs[:, 799],'o') # ++++++
plt.show()
```

```python
fig, axs = plt.subplots(5, 1, figsize=(16,36))

axs[0].plot(range(Lx), vals, '.')
axs[0].set_title('Eigenvalues')

axs[2].plot(range(Lx), vecs[:, 0], 'x')
axs[2].plot(range(Lx), vecs[:, 1], '.')
axs[2].plot(range(Lx), vecs[:, 2],'o')
axs[3].set_title('Eigenvalues')

axs[3].plot(range(Lx), vecs[:, 799],'.r')
axs[3].plot(range(Lx), vecs[:, 798],'ok')
axs[3].set_title('Automatic ticks')


axs[4].plot(range(Lx), vecs[:, 300],'.')
axs[4].plot(range(Lx), vecs[:, 600],'o')
axs[4].set_title('Automatic ticks')

plt.plot()

```
